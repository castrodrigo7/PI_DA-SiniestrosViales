Analizando los datos de la Altura que tiene 567 nulos, procedemos a eliminarla debido a la gran cantidad de faltantes y no existiendo manera de poder rellenar las colummas; ya que altura se refiere a la numeración de la calle donde se produce el siniestro y la mayoria de los casos se producen en cruces o esquinas.

```python
df_hechos= df_hechos.drop("Altura", axis=1) # Se elimina la columna "Altura"
```

Eliminamos la columna Cruce ya que no presenta informacion util para el analisis y tiene 171 datos nulos.

```python
df_hechos= df_hechos.drop("Cruce", axis=1) # Se elimina la columna "Cruce"
```

```python
# Renombro las columnas para normalizar

nuevos_nombres = {
    'Calle': 'CALLE',
    'Dirección Normalizada': 'DIRECCION_NORMALIZADA',
    'pos x': 'LONGITUD',
    'pos y': 'LATITUD',
    }

df_hechos.rename(columns=nuevos_nombres, inplace=True)
```

```python
# Como existe un solo valor nulo lo reemplazamos por la palabra 'SD'
df_hechos.loc[df_hechos['CALLE'].isna(),'CALLE'] = 'SD'
```

```python
df_hechos['DIRECCION_NORMALIZADA'] = df_hechos['DIRECCION_NORMALIZADA'].fillna(df_hechos['LUGAR_DEL_HECHO'])
```

```python
df_hechos.duplicated().sum()
```

No se encontraron datos duplicados

La primera columna con tipo de dato que necesita transformación es HH, debido a que se encuentra como objeto y deberia ser númerico.

```python
df_hechos["HH"].unique()
```

```python
(df_hechos["HH"]=='SD').sum()
```

```python
# Como existe un solo valor 'SD' lo remplazamos en el dataframe original por 12 que es el valor promedio. Y convertimos la columna a tipo entero.
df_hechos['HH'] = df_hechos['HH'].replace('SD', 12)
df_hechos["HH"] = df_hechos["HH"].astype("Int64")
```

```python
df_hechos['LONGITUD'] = pd.to_numeric(df_hechos['LONGITUD'], errors='coerce')  
df_hechos['LATITUD'] = pd.to_numeric(df_hechos['LATITUD'], errors='coerce')
```

```python
df_hechos.loc[df_hechos['LONGITUD'].isna()]
```

```python
#Creo una columna nueva `Día semana` que contenga el nombre del día de la semana
df_hechos['DIA'] = df_hechos['FECHA'].dt.strftime('%A')
df_hechos.insert(8, 'DIA', df_hechos.pop('DIA'))

#Creo un diccionario para modificar a español los nombres de los dás de la semana y lo aplico a la columna nueva
dias={'Monday': 'LUNES', 'Tuesday': 'MARTES', 'Wednesday': 'MIERCOLES',
      'Thursday': 'JUEVES', 'Friday':'VIERNES', 'Saturday': 'SABADO',
       'Sunday': 'DOMINGO'}
df_hechos['DIA'] = df_hechos['DIA'].replace(dias)
```

```python
df_hechos.to_csv('Dataset/df_hechos.csv', index=False)
df_hechos.to_excel('Dataset/df_hechos.xlsx', index=False)
```

```python

# Convertir columnas a datetime
df_victimas['FECHA_FALLECIMIENTO'] = pd.to_datetime(df_victimas['FECHA_FALLECIMIENTO'], errors='coerce')
df_victimas['FECHA'] = pd.to_datetime(df_victimas['FECHA'], errors='coerce')

# Extraer solo la fecha (sin hora)
df_victimas['FECHA'] = df_victimas['FECHA'].dt.date

# Reemplazar valores nulos en 'FECHA_FALLECIMIENTO' con las fechas de 'FECHA'
df_victimas['FECHA_FALLECIMIENTO'] = df_victimas['FECHA_FALLECIMIENTO'].fillna(df_victimas['FECHA'])
```

```python
df_victimas['EDAD'] = pd.to_numeric(df_victimas['EDAD'], errors='coerce').astype('Int64')
df_victimas["EDAD"] = df_victimas["EDAD"].fillna(0)
```

```python
df_victimas ["EDAD"].unique()
```

```python
bins = [0, 1, 20, 40, 60, 80, 120]
labels = ['SD', 'Menos de 20', ' Entre 21-40', 'Entre 41-60', 'Entre 61-80', 'Más de 80']

# Utilizar pd.cut() para crear la columna de rango etario
df_victimas['RANGO_ETARIO'] = pd.cut(df_victimas['EDAD'], bins=bins, labels=labels, right=False)
```

## Carga y union de datos

Cargo los 3 archivos usados para el analisis

```python
df_hechos=pd.read_csv('Dataset/df_hechos.csv')
df_victimas=pd.read_csv('Dataset\df_victimas.csv')
comunas=pd.read_csv('Dataset\comunas.csv')
```

Fusiono los DataFrames df_hechos y df_victimas (con las columnas que considero necesarias a explorar), utilizando la columna 'ID' como clave de fusión. 

```python
df_merge = pd.merge(df_hechos, df_victimas[['ID', 'ROL', 'SEXO', 'EDAD', 'FECHA_FALLECIMIENTO', 'RANGO_ETARIO']], on='ID', how='outer')
```

Vuelvo a unir el DataFrame anterior con el de comunas utilizando la columna 'COMUNA' como clave de fusión

```python
df_merge = pd.merge(df_merge, comunas, on='COMUNA', how='left')
```

## Gráficos de variables cualitativas
### Distribución de variables
Analizaremos ahora las variables categóricas.

### Distribución de los Siniestros según el día de la semana

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='DIA', data=df_merge, order=df_merge['DIA'].value_counts().index, palette='deep')
plt.title(f'Cantidad de víctimas por día de la semana')
plt.ylabel('Día de la semana')
plt.xlabel('Número de víctimas')
plt.show()
```

Conclusión: Si bien los dias sabados y lunes son los mas propensos a accidentes, no existe una gran diferencia entre los dias a considerar.

### Distribución de los Siniestros según el Tipo de Calle

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='TIPO_DE_CALLE', data=df_merge, order=df_merge['TIPO_DE_CALLE'].value_counts().index, palette='deep')
plt.title(f'Cantidad de accidentes según el tipo de calle')
plt.ylabel('Tipo de calle')
plt.xlabel('Número de accidentes')
plt.show()
```

Conclusión: 
Este grafico indica que las avenidas son lugares donde ocurren más siniestros viales en comparación con otros lugares. Las razones posibles incluyen un mayor volumen de tráfico, velocidades promedio más altas, intersecciones complejas, una alta concentración de negocios y actividades, y una mayor cantidad de carriles. Estos factores pueden aumentar la interacción entre vehículos y peatones, lo que a su vez puede aumentar la probabilidad de accidentes. Como análisis subrayo la importancia de implementar medidas de seguridad vial efectivas en las avenidas para prevenir accidentes. 

### Distribución de los Siniestros según los Participantes

```python
plt.figure(figsize=(10, 10))
sns.countplot(y='PARTICIPANTES', data=df_merge, order=df_merge['PARTICIPANTES'].value_counts().index, palette='deep')
plt.title(f'Cantidad siniestros según los participantes')
plt.ylabel('Participantes del Siniestro')
plt.xlabel('Número de víctimas')
plt.show()
```

Conclusión: 

Los sinistros viales mas frecuentes se dan entre Peatones y Pasajeros, que es interpretado como los accidentes entre el peaton y el transporte publico. Seguido por la presencia de accidentes que involucran a las motos

Sugiero que la alta densidad de tráfico, especialmente de colectivos y motos, en la Ciudad Autónoma de Buenos Aires puede estar contribuyendo a un mayor número de siniestros viales. Las interacciones entre peatones, pasajeros y vehículos, especialmente en las paradas de colectivo, pueden ser particularmente riesgosas. Además, el comportamiento de los conductores y la infraestructura vial existente también pueden estar influyendo en la seguridad vial. Resalto la importancia de implementar medidas de seguridad vial efectivas, como una mejor señalización, carriles exclusivos para colectivos, y garantizar la seguridad de los peatones en las paradas de colectivo. También subrayo la necesidad de fomentar el respeto por las normas de tráfico entre todos los usuarios de la vía.

### Distribución de Rol de las víctimas según el Sexo

### Distribución Etaria de Víctimas segun el Sexo

```python
df_hombres = df_merge[(df_merge['SEXO'] == 'MASCULINO') & (df_merge['EDAD'] > 0)]
df_mujeres = df_merge[(df_merge['SEXO'] == 'FEMENINO') & (df_merge['EDAD'] > 0)]

plt.figure(figsize=(12, 6))
sns.histplot(df_hombres['EDAD'], bins=20, kde=True, color='skyblue', label='Hombres')
sns.histplot(df_mujeres['EDAD'], bins=20, kde=True, color='lightcoral', alpha=0.6, label='Mujeres')
plt.title('Distribución Etaria de Víctimas segun el Sexo')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

Conclusión: La imagen muestra que los hombres jóvenes son más propensos a verse involucrados en siniestros viales, con un pico entre los 20 a 40 años. Por otro lado, la distribución de edades en las mujeres no muestra un patrón claro, indicando que la edad no influye significativamente en su propensión a siniestros viales. Esto sugiere que las medidas de prevención y educación vial podrían enfocarse en los hombres jóvenes para reducir la incidencia de accidentes

### Distribución relacionando horas y días, de víctimas fatales

```python
data = df_merge.groupby(['DIA', "HH"])['N_VICTIMAS'].sum().reset_index()

# Crear una tabla pivot para los datos
pivot_data = data.pivot(index='DIA', columns='HH', values='N_VICTIMAS')

# Crear un mapa de calor (heatmap)
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_data, cmap='YlGnBu')
plt.xlabel('Hora del Día')
plt.ylabel('Día de la Semana')
plt.title('Relación entre Número de Víctimas por Horas y Días de la Semana')

plt.show()
```

Conclusión: 

La gráfica muestra que hay momentos específicos durante la semana en los que se registra un mayor número de víctimas. Por ejemplo, hay una concentración más alta de víctimas durante las últimas horas de la noche en ciertos días, lo que podría sugerir períodos de mayor riesgo. Esta información podría ser relevante para medidas de seguridad pública o la asignación de recursos para servicios de emergencia.

Los fines de semana, especialmente durante la noche, es cuando más personas suelen salir a socializar, lo que a menudo implica el consumo de alcohol. Desafortunadamente, algunos pueden tomar la decisión de conducir después de beber, lo que aumenta significativamente el riesgo de siniestros viales. Este comportamiento es especialmente preocupante entre los jóvenes, que pueden carecer de experiencia al volante y subestimar los riesgos del alcohol al conducir.
